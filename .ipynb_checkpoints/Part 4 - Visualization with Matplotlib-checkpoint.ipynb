# Два интерфейса по цене одного

## Интерфейс в стиле MATLAB

```python
plt.figure() # создаем рисунок для графика

# Создаем первую из двух областей графика и задаем текущую ось
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# Создаем вторую область и задаем текущую ось
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

## Объектно-ориентированный интерфейс

```python
# Сначала создаем сетку графиков
# ax будет массивом из двух объектов Axes
fig, ax = plt.subplots(2)

# Вызываем метод plot() соответствующего объекта
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

# Простые линейные графики

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure() # создание рисунка
ax = plt.axes() # создание системы координат

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x)); # plot функция для построения графика данных
```

```python
plt.plot(x, np.sin(x)); # тоже самое через интерфейс pylab
```

```python
plt.plot(x, np.sin(x)) # простой рисунок с несколькими линиями
plt.plot(x, np.cos(x)); # можно несколько раз вызвать функцию plot
```

## Настройка графиков: цвета и стили линий

```python
plt.plot(x, np.sin(x - 0), color='blue') # задаем цвет по названию
plt.plot(x, np.sin(x - 1), color='g') # кратикий код цвета (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Шкала оттенков серого цвета, значение в диапазоне от 0 до 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # 16 разрядный код
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) # кортеж rgb, значения от 0 до 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # поддерживаются все названия цветов html
```

```python
plt.plot(x, x + 0, linestyle='solid') # стиль линий через слово linestyle
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

```python
# можно использовать и следующие сокращенные коды
plt.plot(x, x + 4, linestyle='-') # сплошная линия
plt.plot(x, x + 5, linestyle='--') # штржихованная линия
plt.plot(x, x + 6, linestyle='-.') # штрихпунктирная линия
plt.plot(x, x + 7, linestyle=':'); # пунктирная линия
```

```python
# можно объединить задание кодов linestyle и color в одном неключевом аргументе
plt.plot(x, x + 0, '-g') # сплошная линия зеленого цвета
plt.plot(x, x + 1, '--c') # штрихованная линия голубого цвета
plt.plot(x, x + 2, '-.k') # штрихованная линия черного цвета
plt.plot(x, x + 3, ':r'); # пунктирная линия красного цвета
```

# Простые диаграмы рассеяния

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Построение диаграмм рассеяния с помощью функции plt.plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=f"marker='{marker}'")
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
plt.plot(x, y, '-ok'); # линия (-), маркер круга (о), черный цвет (k)
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

## Построение диаграмм рассеяния с помощью функции plt.scatter

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar(); # отображаем цветовую шкалу
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

# Визуализация погрешностей

## Простые планки погрешностей

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0);
```

## Непрерывные погрешности

```python
from sklearn.gaussian_process import GaussianProcessRegressor

# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the Gaussian process fit
gp = GaussianProcessRegressor(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,
                     random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE)  # 2*sigma ~ 95% confidence region
```

# Графики плотности и контурные графики

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

### визуализация трехмерных функций

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

# Гистрограммы, разбиение по интервалам и плотность

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```python
plt.hist(data);
```

```python
plt.hist(data, bins=30, density=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

## Двумерные гистрограммы и разбиение по интервалам

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

### Функция plt.gist2d: двумерная гистограмма

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('count in bin') # Количество в интервале
```

```python
counts, xedges, yedeges = np.histogram2d(x, y, bins=30)
```

### Функция plt.hexbin: гексагональное разбиение по интервалам

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin') # количество в интервале
```

### Ядерная оценка плотности распределения

```python
from scipy.stats import gaussian_kde

# Выполняем подбор на массиве размера [ndim, nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# вычисляем на регулярной координатной сетке
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Выводим график результата в виде изображения
plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[-3.5, 3.5, -6, 6],
           cmap='Blues')
cb = plt.colorbar()
cb.set_label('dendity') # Плотность
```
